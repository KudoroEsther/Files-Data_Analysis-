```python
import numpy as np
import matplotlib.pyplot as plt
```

### Creating Vectors

### Vector Addition

```python
a = np.array([2, 3])
b = np.array([1, 5])
print(f"a + b = {a + b}")
```

```python
#Vector addition 
a = np.array([1,2,3])
b = np.array([4,5,6])

vector_sum = a + b
print(f"Vector a: {a}")
print(f"Vector b: {b}")
print(f"a + b: {vector_sum}")

#Goemetric interpretation of 1D in 2D view
a_2d = np.array([2,3])
b_2d = np.array([1,2])
sum_2d = a_2d + b_2d
print(a_2d[1])
print(f"2D matrix a + 2D matrix b = \n{sum_2d}")

plt.figure(figsize=(8,6))
plt.quiver(0, 0, a_2d[0], a_2d[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.005)

plt.quiver(0,0, b_2d[0], b_2d[1], angles='xy', scale_units='xy', scale=1, color= 'blue', width= 0.005)

plt.quiver(0,0, sum_2d[0], sum_2d[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005)

plt.quiver(a_2d[0], a_2d[1], b_2d[0], b_2d[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.003, alpha=0.5)

plt.xlim(-1, 4) #represent the scale for the x axis
plt.ylim(-1, 6) #represents the scale for y axis
plt.grid(True)
plt.title("Vector Addition: Parallelogram Rule")
#the order of inputting the legend should align with the order defining using plt.quiver. That is, since a_2d was defined first with the color red, the first entry in plt.legend should be an appropriate name for a_2d, in this case 'vector a'.
plt.legend(['Vector a', 'Vector b', 'Vector c'])

plt.show()
```

```python
#Demonstrating 2D in 2D view
a_twoD = np.array([[2,3], [1,2]])
b_twoD = np.array([[4,5], [2,3]])
sum_2D = a_twoD + b_twoD

plt.figure(figsize=(8, 6))
plt.quiver([0,0], [0,0], a_twoD[0], a_twoD[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.005)
plt.quiver([0,0], [0,0], b_twoD[0], b_twoD[1], angles=30, scale_units='xy', scale=1, color='blue', width=0.005)
plt.quiver([0,0], [0,0], sum_2D[0], sum_2D[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005)
plt.quiver(a_twoD[0], a_twoD[1], b_twoD[0], b_twoD[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.003, alpha=0.5)

plt.xlim(-1,10)
plt.ylim(-1,5)
plt.grid(True)
plt.title("Vector Addition: Parallelogram Rule for 2D View")
plt.legend(['Vector a', 'Vector b', 'a + b'])
plt.show()
```

```python
#Properies of vector addition 
c = np.array([1, -1, 2])

#Commutative property: a + b = b + a
print(f"a + b = {a + b}")
print(f"b + a = {b + a}")
#checks if a+b and b+a are equal and prints True or False
print(f"Commutative: {np.array_equal(a + b, b + a)}")

#Associative property: (a+b)+c = a + (b+c)
left = (a+b)+c
right = a + (b+c)
print(f"(a+b)+c = {left}")
print(f"a+(b+c) = {right}")
print(f"Associative: {np.array_equal(left, right)}")
```

```python
#Scalar multipication
vector = np.array([2, 3, 1])
scalar = 3

scaled_vector = scalar*vector
print(f"Original vector: {vector}")
print(f"Scalar: {scalar}")
print(f"Scaled vector: {scaled_vector}")

#Geometic interpretation
original_2d = np.array([2, 1])
scales = [0.5, 1, 2, -1]

fig = plt.figure(figsize=(10,6))

#the next four lines centerlizes the graph, it's not in the original code along
ax = fig.add_subplot(1,1,1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

colors = ['red', 'blue', 'green', 'purple']

for i, scale in enumerate(scales):
    scaled = scale * original_2d
    plt.quiver(0,0, scaled[0], scaled[1], angles='xy', scale_units='xy', scale=1, color=colors[i], width=0.005, label=f'{scale} * v')

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid(True)
plt.title("Scalar Mulitplication Effects")
plt.legend()
plt.show()
```

```python
#Properties of scalar multiplication
a = np.array([1,2,3])
c1, c2 = 2,3

#ASssociative: c1(c2*a) = (c1*c2)*a
left = c1 * (c2 *a)
right = (c1 * c2)*a
print(f"c1(c2*a) = {left}")
print(f"(c1*c2)*a = {right}")
print(f"Associative: {np.array_equal(left, right)}")

#Distributive over vector addition c(a+b) = ca + cb
left = c1 *(a+b)
right = c1 * a + c1 * b
print(f"c(a + b) = {left}")
print(f"ca + cb = {right}")
print(f"Distributive: {np.array_equal(left, right)}")
```

### The Physcist's Perspective of a vector

```python
#The physicist's view: vectors as arrows in space
vector_a = np.array([3,4])
#this gets the squareroot of the squares of 3 and 4 added together
magnitude = np.linalg.norm(vector_a)
print(f"vector: {vector_a}")
print(f"Magnitude: {magnitude}")

#Direction (unit vector)
unit_vector = vector_a/magnitude
print(f"Unit vector (direction): {unit_vector}")

#Visualization
plt.figure(figsize=(8,6))
plt.quiver(0,0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='red', width= 0.005)
plt.quiver(0,0, unit_vector[0], unit_vector[1], angles='xy', scale_units='xy', scale=1, color='blue', width= 0.003)
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.grid(True)
plt.title("Vector: Magnitude and Direction")
plt.legend(['Orignal Vector', 'Unit Vector'])
plt.show()
```

### The Mathematician Perspective

```python
#Mathematician's view: vectors as abstract objects
#Example 1: Polynomials as vectors
#p(x) = 3x^2 +2x + 1 represented as [1,2,3] (coefficients)
poly1 = np.array([1,2,3]) # 1 +2x + 3x^2
poly2 = np.array([2,-1, 1]) # 2 -x + x^2

#vector addition for polynomials
poly_sum = poly1 + poly2
print(f"Polynomial 1: {poly1}")
print(f"Polynomial 2: {poly2}")
print(f"Sum: {poly_sum}")

#Scalar multiplication for polynomials
scalar = 2
poly_scaled = scalar * poly1
print(f"2 * Polynomial 1: {poly_scaled}")


#Example 2: Functions as variables (function space)
def func1(x):
    return x**2

def func2(x):
    return 2*x

#Function addition (f + g)(x) = f(x) + g(x)
def func_sum(x):
    return func1(x) + func2(x)

#split the values between 0 and 3 into 100
x_vals = np.linspace(0, 3, 100)
plt.figure(figsize=(10, 6))
#this plots x_vals against the square of x_vals
plt.plot(x_vals, func1(x_vals), label='f(x) = x')

#this plots x_vals against values that are twice x_vals with x_vals on the x axis
plt.plot(x_vals, func2(x_vals), label='g(x) = 2x')


plt.plot(x_vals, func_sum(x_vals), label= '(f+g)(x) = x+2x')
plt.legend()
plt.title("Functions as vectors")
plt.grid(True)
plt.show()
```
